See how data is correlated to each other 

```python
numeric_df = df.select_dtypes(include=['float64','int64'])
correlation = numeric_df.corr()
correlation
```

```python
sns.heatmap(correlation,annot=True)
plt.rcParams['figure.figsize']=(10,5)
plt.show()
```

Correlation measures the strength and direction of a linear relationship between two variables.

It is expressed as a value between -1 and +1.

Scale:

+1 → Perfect positive correlation (when one increases, the other always increases).

-1 → Perfect negative correlation (when one increases, the other always decreases).

0 → No linear relationship.

**Grouping** also plays an essential role in EDA as it gives concise overview of vast data.
Here, we consider a use case for presenting data for *CONTINENT That Have Grown Faster*

```python
df.groupby('Continent').mean(numeric_only=True)
```

```python
df2= df.groupby('Continent').mean(numeric_only=True).sort_values(by='2022 Population',ascending=False)
df2
```

```python
df3=df2.transpose()
df3
```

```python
df3.plot()
```

```python
df2.columns
```

```python
df2 = df.groupby('Continent')[['1970 Population', '1980 Population', '1990 Population',
       '2000 Population', '2010 Population', '2015 Population',
       '2020 Population', '2022 Population']].mean(numeric_only=True).sort_values(by='2022 Population',ascending=True)
```

```python
df3=df2.transpose()
```
