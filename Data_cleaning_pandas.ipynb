**DATA CLEANING WITH PANDAS**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel(r"D:\PowerBi\Alex the Analyst\Python\Pandas\data_cleaning_pandas\Customer call List.xlsx")
```

```python
df.head()
```

**Stage 1** - Remove Duplicate

```python
df=df.drop_duplicates()
```

**Stage 2** - Drop Not Useful Column

```python
df=df.drop(columns='Not_Useful_Column')
```

**Stage 3**- Standardize Data with cleaning of columns

**a.** Last_Name cleaning

*Strip()* function = Last_Name column has '_,/.' symbols it is cleaned 

```python
df['Last_Name']=df['Last_Name'].str.strip("123._/")
```

**b.** Phone_Number is cleaned and format to standard from

Regular Expression is used to clean

```python
df['Phone_Number']=df['Phone_Number'].str.replace('[^a-zA-Z0-9]','',regex=True)
```

```python
df['Phone_Number']=df['Phone_Number'].apply(lambda x: str(x))
df['Phone_Number']=df['Phone_Number'].apply(lambda x: x[0:3]+'-'+x[3:6]+'-'+x[6:10])
```

```python
df['Phone_Number']=df['Phone_Number'].str.replace('nan--','')
df['Phone_Number']=df['Phone_Number'].str.replace('Na--','')
```

```python
df
```

**c.** Address Cleaning

We split address into different columns based on *Street,State,ZipCode*

```python
df[['Street','State','Zip_code']]=df['Address'].str.split(',',n=2,expand=True)
```

```python
df
```

**d.** Paynig_customers and Do_not_contact Cleaning

Replace Yes -> Y and No -> N

```python
df['Paying Customer']=df['Paying Customer'].str.replace('Yes','Y')
df['Paying Customer']=df['Paying Customer'].str.replace('No','N')
```

```python
df['Do_Not_Contact']=df['Do_Not_Contact'].str.replace('Yes','Y')
df['Do_Not_Contact']=df['Do_Not_Contact'].str.replace('No','N')
```

```python
df
```

**Stage 4**- Repalce N/a and NAN value across Data Frame

```python
df=df.replace('N/a','')
df=df.fillna('')
```

```python
df
```

**Stage 5** - Generating final List of Customer to call
Do_not_call = N and Phone_number != blank

```python

for x in df.index:
    if df.loc[x,'Do_Not_Contact']=='Y':
        df=df.drop(x,inplace=True)
```

```python
df = df[(df['Phone_Number']!='')]
```
